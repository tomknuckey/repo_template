```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
pdf_all = pd.read_csv("../data/intermediate/national_data.csv")
```

```python


fig = go.Figure()

# Election line on left y-axis
fig.add_trace(
    go.Scatter(
        x=pdf_all["Date"],
        y=pdf_all["Conservative_election"],
        name="Conservative Election",
        mode="lines+markers",
        yaxis="y1"
    )
)

# Polling line on right y-axis
fig.add_trace(
    go.Scatter(
        x=pdf_all["Date"],
        y=pdf_all["Conservative_polling"],
        name="Conservative Polling",
        mode="lines+markers",
        yaxis="y2"
    )
)

# Set layout with two y-axes
fig.update_layout(
    title="Conservative Seats vs Polling Over Time",
    yaxis=dict(
        title="Seats",
        side="left"
    ),
    yaxis2=dict(
        title="Polling %",
        overlaying="y",
        side="right"
    ),
    xaxis=dict(title="Date"),
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

```python
cols = ["Unemployment Rate", "Incumbent_Win", "GDP_YoY"]

# Create normalized versions without altering originals
for c in cols:
    pdf_all[f"{c}_norm"] = (
        (pdf_all[c] - pdf_all[c].min()) /
        (pdf_all[c].max() - pdf_all[c].min())
    )
```

```python


px.line(
    pdf_all,
    x="Date",
    y=[f"{c}_norm" for c in cols],
    title="Normalized Indicators Over Time"
)
```

```python
px.line(pdf_all, x="Date", y=["GDP_YoY", "Inflation_YoY"])
```
